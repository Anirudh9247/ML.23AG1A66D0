```python
  import matplotlib.pyplot as plt
  import numpy as np
  import pandas as pd
```

```python
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("My Line Chart")

x = np.array([1, 5])
y = np.array([10, 100])

plt.plot(x, y)
plt.show()
```

```python
x=[10,20,30,40,50]
y=[111,444,999,1666,2555]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='red',linestyle=':',marker='s',linewidth=3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("line chart")
plt.legend(['my data series'])
plt.grid(True)
plt.show()
```

```python
categories=['X','Y','Z','A','B']
values=[5,4,3,2,1]
plt.figure(figsize=(5,5))
plt.bar(categories,values,color='ORANGE')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('visiualize bar')
plt.show()
```

```python
data = [12, 15, 13, 17, 16, 16, 19, 12, 14, 16, 14, 17, 18, 18, 15, 14]

# Create histogram
plt.hist(data, bins=16, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram Example')
plt.show()
```

```python
x = [10, 20, 30, 40, 50]
y = [50, 70, 40, 60, 80]

# Create scatter plot
plt.scatter(x, y)
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Simple Scatter Plot Example')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Simple Pie Chart Example')
plt.axis('equal')
plt.show()
```

```python
data = [7, 8, 5, 6, 7, 9, 12, 5, 7, 8, 6]
plt.boxplot(data)
plt.title('Simple Box Plot Example')
plt.ylabel('Values')
plt.show()
```

```python
data = np.random.rand(5, 5)

# Create heatmap
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()  # Show color scale
plt.title('Simple Heatmap Example')
plt.show()
```
